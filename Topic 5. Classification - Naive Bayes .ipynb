# Defining Dataset

```python
weather=['Sunny','Sunny','Overcast','Rainy','Rainy','Rainy','Overcast','Sunny','Sunny','Rainy',
         'Sunny','Overcast','Overcast','Rainy']
```

```python
temp=['Hot','Hot','Hot','Mild','Cool','Cool','Cool','Mild','Cool','Mild','Mild','Mild','Hot','Mild']
```

```python
play=['No','No','Yes','Yes','Yes','No','Yes','No','Yes','Yes','Yes','Yes','Yes','No']
```

# Encoding Features

```python
from sklearn import preprocessing 
label_encoder = preprocessing.LabelEncoder()
```

```python
weather_encoded = label_encoder.fit_transform(weather)
```

```python
weather_encoded #Overcast:0, Rainy:1, Sunny:2
```

```python
temp_encoded = label_encoder.fit_transform(temp)
```

```python
temp_encoded # Cool:0, Hot:1,  Mild:2
```

```python
play = label_encoder.fit_transform(play)
```

```python
play
```

# Generating Model

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model = GaussianNB()
```

```python
features = np.array(list(zip(weather_encoded,temp_encoded)))
```

```python
features
```

```python
model.fit(features,play)
```

```python
predicted = model.predict([[2,1]]) #[sunny,Hot] = No
```
