```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
```

# Example 1 - A Simple cost function

$f(x) = x^2 + x + 1$

```python
def f(x):
    return x**2 + x + 1
```

```python
x_1 = np.linspace(start=-3, stop=3, num=100)
```

```python
x_1
```

```python
plt.xlim(-3,3)
plt.ylim(0,8)
plt.xlabel("X",fontsize=16)
plt.ylabel("F(X)",fontsize=16)
plt.plot(x_1,f(x_1))
plt.show()
```

## Slope and Derivates

```python
def df(x):
    return 2*x + 1
```

```python
plt.figure(figsize=[15,5])

#chart 1 : Cost function
plt.subplot(1,2,1)
plt.xlim(-3,3)
plt.ylim(0,8)
plt.xlabel("X",fontsize=16)
plt.ylabel("F(X)",fontsize=16)
plt.plot(x_1,f(x_1))

#Slope of cost function
plt.subplot(1,2,2)
plt.xlim(-2,3)
plt.ylim(-3,6)
plt.plot(x_1,df(x_1))
plt.grid()
plt.show()
```

## Python loops and gradient descent

```python
new_x = 3
previous_x = 0
step_multiplier = 0.1
precision = 0.0001

x_list = []
slope_list = []
for n in range(500):
    previous_x = new_x
    gradient = df(previous_x)
    new_x = previous_x - step_multiplier*gradient
    
    step_size = abs(new_x - previous_x)
    
    x_list.append(new_x)
    slope_list.append(df(new_x))
    if step_size < precision:
        print("Loop ran these many times",n)
        break
print("Local Minimum occurs at : " , new_x)
print("Slope or df(x) at this point : ",df(new_x))
print("f(x) or cost value at this point : ", f(new_x))    
```

```python
plt.figure(figsize=[15,5])

#chart 1 : Cost function
plt.subplot(1,3,1)
plt.xlim(-3,3)
plt.ylim(0,8)
plt.title("Cost function")
plt.xlabel("X",fontsize=16)
plt.ylabel("F(X)",fontsize=16)
plt.plot(x_1,f(x_1))
values = np.array(x_list)
plt.scatter(x_list,f(values),color="red")

#Slope of cost function
plt.subplot(1,3,2)
plt.xlim(-2,3)
plt.ylim(-3,6)
plt.title("Slope of the cost function")
plt.xlabel("X",fontsize=16)
plt.ylabel("dF(x)",fontsize=16)
plt.plot(x_1,df(x_1))
plt.scatter(x_list,slope_list,color="red")

plt.subplot(1,3,3)
plt.xlim(-0.55,-0.2)
plt.ylim(-0.3,0.8)
plt.title("Slope of the cost function(closeup)")
plt.xlabel("X",fontsize=16)
plt.ylabel("dF(x)",fontsize=16)
plt.plot(x_1,df(x_1))
plt.scatter(x_list,slope_list,color="red")
plt.grid()
plt.show()
```
