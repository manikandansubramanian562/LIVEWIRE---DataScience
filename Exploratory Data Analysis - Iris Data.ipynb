Exploratory Data Analysis
IRIS DATA SET

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("iris.csv")
```

```python
#Howmany data-points and features?
df.shape
```

```python
#What are the column names in the data set?
df.columns
```

```python
#Howmany flowers for each species are present?
#Iris data set is s balanced data
df['variety'].value_counts()
```

2-D Scatter plot

```python
df.plot(kind = 'scatter',x = 'sepal.length', y = 'sepal.width');
plt.show()
```

```python
plt.scatter(df['sepal.length'],df['sepal.width'])
plt.show()
```

```python
# Scatter plot with distingushed species
import seaborn as sns
sns.set_style("whitegrid")
sns.FacetGrid(df,hue = "variety",height=4)\
   .map(plt.scatter,"sepal.length","sepal.width")\
   .add_legend()
plt.show()
```

Pair plot

```python
#To visualize 4 Dimensional data(maximum of 6 dimensions)
sns.set_style("whitegrid")
sns.pairplot(df,hue="variety",height=3)
plt.show()
```

```python
#Univariate Analysis (To find Significant features)
sns.set_style("whitegrid")
sns.FacetGrid(df,hue = "variety",height=5)\
   .map(sns.distplot,"sepal.length")\
   .add_legend()
plt.show()
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(df,hue = "variety",height=5)\
   .map(sns.distplot,"sepal.width")\
   .add_legend()
plt.show()
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(df,hue = "variety",height=5)\
   .map(sns.distplot,"petal.length")\
   .add_legend()
plt.show()
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(df,hue = "variety",height=5)\
   .map(sns.distplot,"petal.width")\
   .add_legend()
plt.show()
```

```python
#mean (Petal length - Steosa, Viginica, Versicolor)
print(np.mean(df_setosa["petal.length"]))
print(np.mean(df_Versicolor["petal.length"]))
print(np.mean(df_virginica["petal.length"]))
#Standard deviation(Petal length - Steosa, Viginica, Versicolor)
print(np.std(df_setosa["petal.length"]))
print(np.std(df_Versicolor["petal.length"]))
print(np.std(df_virginica["petal.length"]))
```

```python
#median, Quantile, Percentile
print("median")
print(np.median(df_setosa['petal.length']))
print(np.median(df_Versicolor['petal.length']))
print(np.median(df_virginica['petal.length']))
```

```python
print("Percentile")
print(np.percentile(df_setosa['petal.length'],np.arange(0,100,25)))
print(np.percentile(df_Versicolor['petal.length'],np.arange(0,100,25)))
print(np.percentile(df_virginica['petal.length'],np.arange(0,100,25)))
```
