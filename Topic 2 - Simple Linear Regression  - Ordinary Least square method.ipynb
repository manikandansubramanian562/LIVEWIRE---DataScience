```python
import numpy as np #Import libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Salary_Data.csv") #Import data
```

```python
X = df["YearsExperience"].values 
Y = df["Salary"].values
```

```python
x_mean = np.mean(X)
y_mean = np.mean(Y)
```

```python
num = 0
denom = 0
```

```python
from IPython.display import Image
Image("ols.png")
```

```python
m = len(X)
for i in range(m):
    num += (X[i] - x_mean)*(Y[i] - y_mean)
    denom += (X[i] - x_mean)**2
m = num / denom
c = y_mean - m * x_mean
```

```python
print(m,c)
y = m*X+c
```

```python
plt.scatter(X,Y,color = 'green',label="Scatter plot")
plt.plot(X,y,color = "red",label = "Regression Line")
plt.xlabel("Years of experience")
plt.ylabel("Salary")
plt.legend()
plt.show()
```
